1. Create a `matrix` class with the following properties:
    * It can be initialized in 2 ways:
        1. with arguments `n` and `m`, the size of the matrix. A newly instanciated matrix will contain all zeros.
        2. with a list of lists of values. Note that since we are using lists of lists to implement matrices, it is possible that not all rows have the same number of columns. Test explicitly that the matrix is properly specified.
    * Matrix instances `M` can be indexed with `M[i][j]` and `M[i,j]`.
    * Matrix assignment works in 2 ways:
        1. If `M_1` and `M_2` are `matrix` instances `M_1=M_2` sets the values of `M_1` to those of `M_2`, if they are the same size. Error otherwise.
        2. In example above `M_2` can be a list of lists of correct size.


```python
class Matrix:
    def __init__(self, n, m=None):
        if isinstance(n, list):  # Initialize with a list of lists
            self.data = n
            self.n = len(n)
            self.m = len(n[0]) if n else 0
            if any(len(row) != self.m for row in n):
                raise ValueError("Error: All rows need to have the same number of columns")
        else:  # Initialize with dimensions
            self.n = n
            self.m = m
            self.data = [[0] * m for _ in range(n)]
    
    def __getitem__(self, indices):
        if isinstance(indices, tuple):
            i, j = indices
            return self.data[i][j]
        return self.data[indices]
    
    def __setitem__(self, indices, value):
        if isinstance(indices, tuple):
            i, j = indices
            self.data[i][j] = value
        else:
            self.data[indices] = value
    
    def __repr__(self):
        return "\n".join(" ".join(map(str, row)) for row in self.data)
    
    def assign(self, other):
        if isinstance(other, Matrix):
            if self.n != other.n or self.m != other.m:
                raise ValueError("Error: Matrices must have the same dimensions for assignment")
            self.data = [row[:] for row in other.data]
        elif isinstance(other, list):
            if len(other) != self.n or any(len(row) != self.m for row in other):
                raise ValueError("Error: Dimensions must match for assignment")
            self.data = [row[:] for row in other]
        else:
            raise TypeError("Error: Assignment must be from a Matrix or list of lists")
```

5. Overload python operators to appropriately use your functions in 4 and allow expressions like:
    * 2*M
    * M*2
    * M+N
    * M-N
    * M*N
    * M==N
    * M=N


```python
class Matrix:
    def __init__(self, n, m=None):
        if isinstance(n, list):  
            self.data = [row[:] for row in n]
            self.n = len(n)
            self.m = len(n[0]) if n else 0
            if any(len(row) != self.m for row in n):
                raise ValueError("Error: All rows need to have the same number of columns")
        else:  
            self.n = n
            self.m = m
            self.data = [[0] * m for _ in range(n)]
    
    def __repr__(self):
        return "\n".join(" ".join(map(str, row)) for row in self.data)
    
    def __mul__(self, other):
        if isinstance(other, (int, float)):  # Scalar multiplication
            return self.scalarmul(other)
        elif isinstance(other, Matrix):  # Matrix multiplication
            return self.mat_mult(other)
        else:
            raise TypeError("Unsupported operand type for *: Matrix and {}".format(type(other)))
    
    def __rmul__(self, other):
        return self.__mul__(other)
    
    def __add__(self, other):
        return self.add(other)
    
    def __sub__(self, other):
        return self.sub(other)
    
    def __eq__(self, other):
        return self.equals(other)
    
    def __copy__(self):
        return Matrix([row[:] for row in self.data])
    
    def __assign__(self, other):
        if not isinstance(other, Matrix):
            raise TypeError("Can only assign another Matrix")
        self.data = [row[:] for row in other.data]
        self.n, self.m = other.shape()
    
    def shape(self):
        return (self.n, self.m)
    
    def scalarmul(self, c):
        return Matrix([[c * val for val in row] for row in self.data])
    
    def add(self, N):
        return Matrix([[self.data[i][j] + N.data[i][j] for j in range(self.m)] for i in range(self.n)])
    
    def sub(self, N):
        return Matrix([[self.data[i][j] - N.data[i][j] for j in range(self.m)] for i in range(self.n)])
    
    def mat_mult(self, N):
        return Matrix([[sum(self.data[i][k] * N.data[k][j] for k in range(self.m)) for j in range(N.m)] for i in range(self.n)])
    
    def equals(self, N):
        return self.data == N.data

```

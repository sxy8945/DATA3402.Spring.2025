1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:
    def __init__(self, max_value):
        self.max_value=max_value
        self.count=0
        
    def increment(self):
        if self.count<self.max_value:
            self.count+=1
        else:
            print(f"Error: cannot increment past {self.max_value}")
            
    def reset(self):
        self.count=0
        
    def __str__(self):
        return f"value={self.count}"
```

7. Add a function to the object classes, including the base, that returns a list of up to 16 pairs of  $x$ and $y$ points on the parameter of the object. 

```python
import math

# Base Class for Shapes
class Base:
    def get_perimeter_points(self):
        raise NotImplementedError("Method should be overridden by subclasses.")

# Rectangle Class
class Rectangle(Base):
    def __init__(self, x, y, width, length):
        self.x = x
        self.y = y
        self.width = width
        self.length = length

    def get_perimeter_points(self):
        #Returns evenly spaced perimeter points around the rectangle.
        points = []
        step_w = self.width / 4
        step_l = self.length / 4

        # Top Edge
        points += [(self.x + i * step_w, self.y) for i in range(5)]
        # Right Edge
        points += [(self.x + self.width, self.y + i * step_l) for i in range(1, 5)]
        # Bottom Edge
        points += [(self.x + self.width - i * step_w, self.y + self.length) for i in range(1, 5)]
        # Left Edge
        points += [(self.x, self.y + self.length - i * step_l) for i in range(1, 4)]

        return points[:16]

# Circle Class
class Circle(Base):
    def __init__(self, x, y, radius):
        self.x = x
        self.y = y
        self.radius = radius

    def get_perimeter_points(self):
        #Returns 16 evenly spaced perimeter points around the circle.
        points = []
        for i in range(16):
            angle = (2 * math.pi / 16) * i
            px = self.x + self.radius * math.cos(angle)
            py = self.y + self.radius * math.sin(angle)
            points.append((px, py))
        return points

# Triangle Class
class Triangle(Base):

    def __init__(self, x1, y1, x2, y2, x3, y3):
        self.x1, self.y1 = x1, y1
        self.x2, self.y2 = x2, y2
        self.x3, self.y3 = x3, y3

    def get_perimeter_points(self):
        #Returns 16 evenly spaced perimeter points along the triangle edges.
        def get_points_between(x_start, y_start, x_end, y_end, n=5):
            #Calculates `n` evenly spaced points between two given points.
            return [
                (x_start + i * (x_end - x_start) / n, y_start + i * (y_end - y_start) / n)
                for i in range(n)
            ]

        points = []
        points += get_points_between(self.x1, self.y1, self.x2, self.y2)
        points += get_points_between(self.x2, self.y2, self.x3, self.y3)
        points += get_points_between(self.x3, self.y3, self.x1, self.y1)

        return points[:16]



```

8. Add a function to the object classes, including the base, that tests if a given set of $x$ and $y$ coordinates are inside of the object. You'll have to think through how to determine if a set of coordinates are inside an object for each object type.

```python
import math

class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        self._length = length
        self._width = width
        self._x = x
        self._y = y

    def area(self):
        return self._length * self._width

    def perimeter(self):
        return 2 * (self._length + self._width)

    def get_coordinates(self):
        return (self._x, self._y)

    def perimeter_points(self, num_points=16):
        points = []
        pts_per_side = num_points // 4  # points on each side
        for i in range(num_points):
            if i < pts_per_side:
                # bottom side: from (x, y) to (x+length, y)
                t = i / pts_per_side
                x_val = self._x + t * self._length
                y_val = self._y
            elif i < 2 * pts_per_side:
                # right side:
                t = (i - pts_per_side) / pts_per_side
                x_val = self._x + self._length
                y_val = self._y + t * self._width
            elif i < 3 * pts_per_side:
                # top side
                t = (i - 2 * pts_per_side) / pts_per_side
                x_val = self._x + self._length - t * self._length
                y_val = self._y + self._width
            else:
                # left side
                t = (i - 3 * pts_per_side) / pts_per_side
                x_val = self._x
                y_val = self._y + self._width - t * self._width
            points.append((x_val, y_val))
        return points

    def is_inside(self, x, y):
        # For an axis-aligned rectangle:
        return (self._x <= x <= self._x + self._length) and (self._y <= y <= self._y + self._width)


class Circle(Shape):
    def __init__(self, radius, x, y):
        self._radius = radius
        self._x = x
        self._y = y

    def area(self):
        return math.pi * (self._radius ** 2)

    def perimeter(self):
        return 2 * math.pi * self._radius
        
    def get_coordinates(self):
        return (self._x, self._y)
    
    def perimeter_points(self, num_points=16):
        points = []
        for i in range(num_points):
            angle = 2 * math.pi * i / num_points
            x_val = self._x + self._radius * math.cos(angle)
            y_val = self._y + self._radius * math.sin(angle)
            points.append((x_val, y_val))
        return points

    def is_inside(self, x, y):
        # Check if (x, y) is within the circle's radius
        return ((x - self._x) ** 2 + (y - self._y) ** 2) <= self._radius ** 2


class Triangle(Shape):
    def __init__(self, side1, side2, side3, x, y):
        self._side1 = side1
        self._side2 = side2
        self._side3 = side3
        self._x = x
        self._y = y

    def area(self):
        s = (self._side1 + self._side2 + self._side3) / 2
        return math.sqrt(s * (s - self._side1) * (s - self._side2) * (s - self._side3))

    def perimeter(self):
        return self._side1 + self._side2 + self._side3

    def get_coordinates(self):
        return (self._x, self._y)

    def get_sides(self):
        return (self._side1, self._side2, self._side3)

    def perimeter_points(self, num_points=16):
        points = []
        pts_per_side = num_points // 3  # points per side for the triangle
        for i in range(pts_per_side):
            # side 1
            t = i / pts_per_side
            x_val = self._x + t * self._side1
            y_val = self._y
            points.append((x_val, y_val))
    
            # side 2
            t = i / pts_per_side
            x_val = self._x + self._side1
            y_val = self._y + t * self._side2
            points.append((x_val, y_val))
    
            # side 3
            t = i / pts_per_side
            x_val = self._x + self._side1 - t * self._side1
            y_val = self._y + self._side2
            points.append((x_val, y_val))
        return points[:num_points]

    def is_inside(self, x, y):

        A = (self._x, self._y)
        B = (self._x + self._side1, self._y)
        C = (self._x + self._side1, self._y + self._side2)
        
        def sign(p1, p2, p3):
            return (p1[0] - p3[0]) * (p2[1] - p3[1]) - (p2[0] - p3[0]) * (p1[1] - p3[1])
        
        d1 = sign((x, y), A, B)
        d2 = sign((x, y), B, C)
        d3 = sign((x, y), C, A)
        
        neg = (d1 < 0) or (d2 < 0) or (d3 < 0)
        pos = (d1 > 0) or (d2 > 0) or (d3 > 0)
        return not (neg and pos)
```

9. Add a function in the base class of the object classes that returns true/false testing that the object overlaps with another object.

```python
import math

class Base(object):
    def area(self):
        return None

    def perimeter(self):
        return None

    def check_overlap(self, other):
        return None

class Circle(Base):
    def __init__(self, x, y, r):
        self.x = x
        self.y = y
        self._radius = r

    def area(self):
        return math.pi * self._radius ** 2

    def perimeter(self):
        return 2 * math.pi * self._radius

    def check_overlap(self, other):
        if isinstance(other, Circle):
            dist_sq = (self.x - other.x) ** 2 + (self.y - other.y) ** 2
            rad_sum_sq = (self._radius + other._radius) ** 2
            return dist_sq <= rad_sum_sq
        return False

class Rectangle(Base):
    def __init__(self, x, y, width, length):
        self.x = x
        self.y = y
        self._width = width
        self._length = length

    def area(self):
        return self._width * self._length

    def perimeter(self):
        return 2 * (self._width + self._length)

    def check_overlap(self, other):
        if isinstance(other, Rectangle):
            # Check if rectangles do NOT overlap:
            if (self.x + self._width <= other.x or 
                other.x + other._width <= self.x or
                self.y + self._length <= other.y or 
                other.y + other._length <= self.y):
                return False
            return True
        return False
```

```python
#Canvas Class, (importing class from paint module did not work)
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.data = [[' '] * width for _ in range(height)]

    def set_pixel(self, row, col, **kargs):
        char = kargs.get('char', '*')
        if 0 <= row < self.height and 0 <= col < self.width:
            self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]

    def clear_canvas(self):
        self.data = [[' '] * self.width for _ in range(self.height)]

    def v_line(self, x, y, w, **kargs):
        for i in range(x, x + w):
            self.set_pixel(i, y, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(y, y + h):
            self.set_pixel(x, i, **kargs)

    def line(self, x1, y1, x2, y2, **kargs):
        char = kargs.get('char', '*')
        dx = abs(x2 - x1)
        dy = abs(y2 - y1)
        sx = 1 if x1 < x2 else -1
        sy = 1 if y1 < y2 else -1
        err = dx - dy

        while True:
            self.set_pixel(x1, y1, char=char)
            if x1 == x2 and y1 == y2:
                break
            e2 = 2 * err
            if e2 > -dy:
                err -= dy
                x1 += sx
            if e2 < dx:
                err += dx
                y1 += sy

    def display(self):
        print("\n".join(["".join(row) for row in self.data]))


# Rectangle class 
class Rectangle:
    def __init__(self, x, y, width, height, **kargs):
        self.x = x
        self.y = y
        self.width = width
        self.height = height
        self.kargs = kargs

    def draw(self, canvas, **kargs):
        draw_kargs = {**self.kargs, **kargs, "char": "*"}  
        canvas.h_line(self.x, self.y, self.width, **draw_kargs)
        canvas.h_line(self.x + self.height - 1, self.y, self.width, **draw_kargs)
        canvas.v_line(self.x, self.y, self.height, **draw_kargs)
        canvas.v_line(self.x, self.y + self.width - 1, self.height, **draw_kargs)


# Triangle class 
class Triangle:
    def __init__(self, x, y, height, **kargs):
        self.x = x  
        self.y = y  
        self.height = height
        self.kargs = kargs

    def draw(self, canvas, **kargs):
        draw_kargs = {**self.kargs, **kargs, "char": "^"}  

        # Draw the triangle sides
        for i in range(self.height):
            left_x = self.x - i  # Left edge of the triangle
            right_x = self.x + i  # Right edge of the triangle
            y_pos = self.y + i  # Current row
            canvas.set_pixel(y_pos, left_x, **draw_kargs)
            canvas.set_pixel(y_pos, right_x, **draw_kargs)

        # base of the triangle
        canvas.h_line(self.y + self.height - 1, self.x - self.height + 1, 2 * self.height - 1, **draw_kargs)


# CompoundShape class 
class CompoundShape:
    def __init__(self):
        self.shapes = []

    def add_shape(self, shape):
        self.shapes.append(shape)

    def draw(self, canvas, **kargs):
        for shape in self.shapes:
            shape.draw(canvas, **kargs)
```

```python
class Shape:
    
    def draw(self, canvas):
        pass

    def modify(self, x, y, size):
        pass


class Rectangle(Shape):
    def __init__(self, x, y, width, height):
        self.x = x
        self.y = y
        self.width = width
        self.height = height

    def draw(self, canvas):
        for i in range(self.y, self.y + self.height):
            for j in range(self.x, self.x + self.width):
                canvas[i][j] = "*"

    def modify(self, x, y, width, height):   #modifies rectangle position and sizee
        self.x = x
        self.y = y
        self.width = width
        self.height = height


class Triangle(Shape):
    def __init__(self, x, y, size):
        self.x = x
        self.y = y
        self.size = size

    def draw(self, canvas):
        for i in range(self.size):
            for j in range(-i, i + 1):
                canvas[self.y + i][self.x + j] = "*"

    def modify(self, x, y, size):
        self.x = x
        self.y = y
        self.size = size


class RasterDrawing:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.shapes = []

    def add_shape(self, shape):
        self.shapes.append(shape)

    def modify_rectangle(self, index, x, y, width, height):  #modifies the shapes properties if it exists in the list
        if isinstance(self.shapes[index], Rectangle):      
            self.shapes[index].modify(x, y, width, height)

    def modify_triangle(self, index, x, y, size):
        if isinstance(self.shapes[index], Triangle):
            self.shapes[index].modify(x, y, size)

    def paint(self):
        canvas = [[" " for _ in range(self.width)] for _ in range(self.height)]

        for shape in self.shapes:
            shape.draw(canvas)

        for row in canvas:
            print("".join(row))
        print("\n" + "=" * self.width + "\n")  
```

```python

class Shape:
    def __init__(self, name=""):
        self.name = name

    def paint(self, canvas):
        pass

    def __repr__(self):   #str for saving the shape
        return "Shape(" + repr(self.name) + ")"


# Rectangle Class
class Rectangle(Shape):
    def __init__(self, x, y, w, h, name=""):
        Shape.__init__(self, name)
        self.x, self.y, self.w, self.h = x, y, w, h

    def paint(self, canvas):
        for i in range(self.h):
            for j in range(self.w):
                canvas.set_pixel(self.y + i, self.x + j)

    def __repr__(self):
        return f"Rectangle({self.x}, {self.y}, {self.w}, {self.h}, {repr(self.name)})"


# Triangle Class
class Triangle(Shape):
    def __init__(self, x, y, height, name=""):  #isosceles triangle at (x, y) with given height
        Shape.__init__(self, name)
        self.x, self.y, self.height = x, y, height

    def paint(self, canvas):
        for i in range(self.height):
            for j in range(-i, i + 1):
                canvas.set_pixel(self.y + i, self.x + j)

    def __repr__(self):
        return f"Triangle({self.x}, {self.y}, {self.height}, {repr(self.name)})"


# Canvas Class
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.data = [[' '] * width for _ in range(height)]

    def set_pixel(self, row, col):
        if 0 <= row < self.height and 0 <= col < self.width:
            self.data[row][col] = "*"            # marks a pixel position with '*'

    def clear(self):
        self.data = [[' '] * self.width for _ in range(self.height)] #clears canvas by emptying all pixels

    def display(self):
        for row in self.data:
            print("".join(row))
        print("=" * self.width)


# RasterDrawing Class
class RasterDrawing:
    def __init__(self):
        self.shapes = dict()
        self.shape_names = list()

    def add_shape(self, shape):
        if shape.name == "":
            shape.name = self.assign_name()
        self.shapes[shape.name] = shape
        self.shape_names.append(shape.name)

    def paint(self, canvas):
        canvas.clear()
        for shape_name in self.shape_names:
            self.shapes[shape_name].paint(canvas)
        canvas.display()

    def assign_name(self):         #Generates a unique name for a shape
        name_base = "shape"
        name = name_base + "_0"
        i = 1
        while name in self.shapes:
            name = name_base + "_" + str(i)
            i += 1
        return name

    def save(self, filename):
        with open(filename, "w") as f:
            for shape in self.shapes.values():
                f.write(shape.__repr__() + "\n")
        print(f"Drawing saved to {filename}")


# Shape Loader Function
def shape_loader(filename):          # loads shapes from a file and recreates them using eval().
    with open(filename, "r") as f:
        return [eval(line.strip()) for line in f]


```

```python
rd = RasterDrawing()

rd.add_shape(Rectangle(10, 2, 8, 3))  # Top rectangle
rd.add_shape(Triangle(18, 7, 4))  # Middle triangle
rd.add_shape(Rectangle(14, 12, 6, 3))  # Bottom rectangle

print("============ Saving Shapes ===========")
rd.save("shapes.txt")
for shape_name in rd.shape_names:
    print(rd.shapes[shape_name])

# Load shapes from the saved file
print("\n============ Loading Shapes ===========")
loaded_shapes = shape_loader("shapes.txt")
for shape in loaded_shapes:
    print(shape)

# adding loaded shapes
rd_new = RasterDrawing()
for shape in loaded_shapes:
    rd_new.add_shape(shape)

```
